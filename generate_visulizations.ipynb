```python
from blackjack import Blackjack
from agent import QLearningAgent, CardCountingAgent, RandomAgent
from main import play_multiple_rounds, play_round_demo
from visualizations import bar_plot_wins, epoch_bar_comparison
```

```python
bj = Blackjack()
```

```python
player_only_ql_agents = []
for epochs in [10,100,1000,10000,100000]:
    qla = QLearningAgent(0.1, 0.1, 0.6, False)
    qla.train(bj, epochs)
    player_only_ql_agents.append(qla)
both_ql_agents = []
for epochs in [10,100,1000,10000,100000]:
    qla = QLearningAgent(0.4, 0.4, 0.5, True)
    qla.train(bj, epochs)
    both_ql_agents.append(qla)
```

```python
for i, epochs in enumerate([10,100,1000,10000,100000]):
    player_only_ql_agents[i].export_table(f"./table/{epochs}_player_only")
    both_ql_agents[i].export_table(f"./table/{epochs}_both")
```

```python


bar_plot_wins(50, ("QLearning Agent(Player Hand Only)\n100000 Epochs", "QLearning Agent(All Known Cards)\n100000 Epochs", "Random Agent", "Card Counting Agent"), (play_multiple_rounds(50, bj, player_only_ql_agents[4]), play_multiple_rounds(50, bj, both_ql_agents[4]), play_multiple_rounds(50, bj, RandomAgent()),play_multiple_rounds(50, bj, CardCountingAgent())))
```

```python
epoch_bar_comparison(50, (10,100,1000,10000,100000), (play_multiple_rounds(50, bj, player_only_ql_agents[0]), play_multiple_rounds(50, bj, player_only_ql_agents[1]), play_multiple_rounds(50, bj, player_only_ql_agents[2]), play_multiple_rounds(50, bj, player_only_ql_agents[3]), play_multiple_rounds(50, bj, player_only_ql_agents[4])))
```

```python
epoch_bar_comparison(50, bj, (10,100,1000,10000,100000), (play_multiple_rounds(50, bj, both_ql_agents[0]), play_multiple_rounds(50, bj, both_ql_agents[1]), play_multiple_rounds(50, bj, both_ql_agents[2]), play_multiple_rounds(50, bj, both_ql_agents[3]), play_multiple_rounds(50, bj, both_ql_agents[4])))
```
